```python
#Which are the 10 best-selling tracks?

top_tracks = pd.read_sql_query("""
SELECT 
    t.Name AS Track,
    SUM(il.Quantity) AS TotalSold
FROM InvoiceLine il
JOIN Track t 
ON il.TrackId = t.TrackId
GROUP BY t.Name
ORDER BY TotalSold DESC
LIMIT 10;
""", conn)  

top_tracks
```

```python
#Which country generates the most revenue?

country_sales = pd.read_sql_query("""
SELECT 
    c.Country,
    ROUND(SUM(i.Total), 2) AS Revenue
FROM Customer c
JOIN Invoice i
ON c.CustomerId = i.CustomerId
GROUP BY c.Country
ORDER BY Revenue DESC;
""", conn)  

country_sales
```

```python
# Who is the top-performing sales employee?

top_employee = pd.read_sql_query("""
SELECT 
    e.FirstName || ' ' || e.LastName AS Employee,
    ROUND(SUM(i.Total), 2) AS Revenue
FROM Employee e
JOIN Customer c
ON e.EmployeeId = c.SupportRepId
JOIN Invoice i
ON c.CustomerId = i.CustomerId
GROUP BY Employee
ORDER BY Revenue DESC;  
""", conn)

top_employee
```
